```python
#prime numbers
num = int(input("enter the number"))
count = 0
for i in range(1,num+1):
  if(num%i==0):
    count+=1
if(count==2):
  print("it is a prime")
else:
  print("not a prime")
```

```python
#factorial
digit = int(input("enter the number"))
fact = 1
for i in range(1,digit+1):
  fact = fact*i
  print(fact)
```

```python
#palindrome
num = int(input("enter the number"))
rev = 0
temp = num
while num>0:
  rem = num%10
  rev = rev*10+rem
  num = num//10
  print(rev)
if(temp==rev):
  print("It is a palindrome")
else:
  print("not a palindrome")
```

```python
#armsrtong
num = int(input("enter the number"))
sum = 0
temp = num
l = len(str(num))
for i in range(l+1):
  rem = num%10
  sum = sum +(rem**1)
  num = num//10
print(sum)
if(temp==sum):
  print("it is a armstrong")
else:
  print("not a armstrong")
```

```python
num = int(input("Enter a number: "))
sum = 0

while num > 0:
    digit = num % 10
    sum = sum + digit
    num = num // 10

print("Sum of digits:", sum)
```
