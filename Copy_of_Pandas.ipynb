```python
import numpy as np
import pandas as pd
```

```python
s=pd.Series([1,2,3,4,5,6,7,8])
print(s)
```

```python
print(s.describe())
```

```python
print(s.mean())
print(s.min())
print(s.max())
print(s.std())
```

Data Manipulation

```python
#map(function)
mapped=s.map(lambda x:x**2)
print(mapped)
```

```python
#apply(function)
applied=s.apply(lambda x:x*2)
print(applied)
```

```python
#sort_values()
sorted=s.sort_values()
print(sorted)
```

```python
s1=pd.Series([1,5,6,4,9,22,13,11])
print(s1)
sorted=s1.sort_values()
print(sorted)
```

```python
s1.drop(5)
```

Handling Missing Data

```python
#isnull()
s2=pd.Series([1,12,14,np.nan,1.1,16,76])
print(s2)
s2.isnull()
```

```python
#notnull()
s2.notnull()
```

```python
#fillna
s2.fillna(0)
```

```python
#dropna()
s2.dropna()
```

Indexing,Slicing and Filtering

```python
#iloc()
print(s.iloc[3])
print(s1.iloc[-4])
print(s2.iloc[5 & -2])
```

```python
#loc()
print(s2.loc[5])
```

```python
#(s>n)
```

Aggregation

```python
#Sum()
print(s.sum())
print(s1.sum())
print(s2.sum())
```

```python
#cumsum()
print(s.cumsum())
print(s1.cumsum())
print(s2.cumsum())
```

```python
#aggregate()
print(s.aggregate(['mean','std','sum','min','max']))
print(s1.aggregate(['mean','std','sum','min','max']))
print(s2.aggregate(['mean','std','sum','min','max']))
```
