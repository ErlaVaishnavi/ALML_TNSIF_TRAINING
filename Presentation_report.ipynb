```python
#importing Libraries

import numpy as np
import pandas as pd
import plotly.express as px
```

```python
#loading dataset
df= pd.read_excel("/content/Presentation-TASK (1).xlsx")
```

```python

df.describe()
```

```python
df.head()

df.info()
```

```python
# Check missing values
print(df.isnull().sum())

# Add a check to re-load df if it's empty, ensuring continuity
# This addresses cases where df might become empty due to environment resets or unintended executions.
if df.empty:
    print("DataFrame is empty, re-loading data to ensure processing continues.")
    df = pd.read_excel("/content/Presentation-TASK (1).xlsx")

# Convert 'Technical knowledge' to numeric and fill NaNs with 0, ensuring float type
df['Technical knowledge'] = pd.to_numeric(df['Technical knowledge'], errors='coerce').fillna(0).astype(float)

# Fill NaN values in 'Communication' and 'Attrire' with 0 and ensure float type
df['Communication'] = df['Communication'].fillna(0).astype(float)
df['Attrire'] = df['Attrire'].fillna(0).astype(float)

# IMPORTANT: Removed df.dropna() as it was making the DataFrame empty.
# If you need to drop rows with NaNs in other specific columns, use df.dropna(subset=['column_name']) for targeted removal.

# Check for duplicate records (Do NOT remove duplicates)
print("Duplicate Records:", df.duplicated().sum())

# Create Total Marks column
# All components are now numeric (float), so 'Total Marks' will be numeric (float).
df["Total Marks"] = df["Technical knowledge"] + df["Communication"] + df["Attrire"]

# Verify the updated dataset
print(df.head())
```

```python
# Top 5 by Total Marks
top_total = df.nlargest(5, "Total Marks")
print("\nTop 5 by Total Marks")
print(top_total[['Participate Name','Total Marks']])

# Top 5 in Technical knowledge
top_technical = df.nlargest(5, "Technical knowledge")
print("\nTop 5 in Technical")
display(top_technical[['Participant Name','Technical knowledge']])

# Top 5 in Communication
top_communication = df.nlargest(5, "Communication")
print("\nTop 5 in Communication")
print(top_communication[['Participate Name','Communication']])

# Top 5 in Attrire
top_attrire = df.nlargest(5, "Attrire")
print("\nTop 5 in Attrire")
print(top_attrire[['Participate Name','Attrire']])
```

```python
# Visualization for top 5 students
top_total = df.nlargest(5, "Total Marks")

fig = px.bar(
    top_total,
    x="Participant Name",
    y="Total Marks",
    color="Total Marks",
    text="Total Marks",
    title="Top 5 Students by Total Marks"
)

fig.update_traces(textposition="outside")
fig.show()
```
