```python
#Function with parameters
def greet(name):
  print("Welcome", name)

greet("Vyshu")
```

```python
#Functions without parameters and without return value
def display():
    print("Learning Python Functions")

display()
```

```python
#Function with return value
def add(a,b):
  return a+b
result=add(10,15)
print(result)

```

```python
#function with parameter and return value
def square(num):
  return num*num

result=square(10)
print(result)
```

```python
#Lambda function
add=lambda a,b:a+b
print(add(2,7))
```

```python
square=lambda num:num*num
print(square(6))
```
