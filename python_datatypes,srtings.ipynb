```python
#List
stud = ['vyshu','satwika','prathyusha','anil','raj']
print(stud)
print(type(stud))
stud.append("shila")
print(stud)
stud.insert(1,'naresh')
print(stud)
stud.remove('shila')
print(stud)
stud.pop(0)
print(stud)
stud.sort()
print(stud)
stud.reverse()
print(stud)
l = len(stud)
print(l)
stud = "prathyusha"
stud.count('a')
```

```python
#Tuple
car_brands = ('benz','Thar','suzuki','Audi','Thar')
print(car_brands)
print(len(car_brands))
print(car_brands.count('Thar'))
print(car_brands[2])
```

```python
#set
name1 = {'sravya','shiva','teja','priya','Ajith'}
name2 = {'teja','priya','mouni','micky','raj'}
print(name1)
print(name2)
name1.add('task')
print(name1)
name1.remove('Ajith')
print(name1)
unions = name1.union(name2)
print(unions)
inter = name1.intersection(name2)
print(inter)
dif = name1.difference(name2)
print(dif)
name1.clear()
print(name1)
```

```python
#Dictionary
student = {
    "name": "vyshu",
    "age": 21,
    "course": "AIML"
}
print(student)
print(student.keys())
print(student.values())
print(student.items())
print(student.get('name'))
student.pop('name')
print(student)
student.update({'name':'sweety'})
print(student)
student.update({'age':22})
print(student)
```
