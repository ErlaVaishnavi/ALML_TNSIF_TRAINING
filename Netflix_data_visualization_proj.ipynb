```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("/content/netflix_titles.csv")
df.info()
df.head()
```

```python
data=df.dropna()
data.head()
```

```python
#pie plot
type_counts=data["type"].value_counts()
plt.figure(figsize=(5,5))
plt.pie(type_counts,labels=type_counts.index,autopct='%1.1f%%')
plt.title("content analysis")
plt.show()
```

```python
#bar plot
countries_df = data['country'].dropna().str.split(',').explode()
top_countries=countries_df.value_counts().head(40)
plt.figure(figsize=(10,10))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 40 countries contributing contents to Netflix")
plt.show()
```

```python
#bar plot
directors_df = data['director'].dropna().str.split(',').explode()
top_directors=directors_df.value_counts().head(40)
plt.figure(figsize=(10,10))
sns.barplot(x=top_directors.values, y=top_directors.index)
plt.title("Top 40 titles of Directors")
plt.show()
```
